```python
#zad2 
lista1 = ["Patryk", "Ala", "Jacek"]
lista2 = []

print(lista1)
print(lista2)

while len(lista1) != 0:
  for i in lista1:
    lista2.append(i)
    lista1.remove(i)

print(lista1)
print(lista2)
```

```python
#zad2 
lista1 = []
lista2 = ["Patryk", "Ala","Jacek"]

print(lista1)
print(lista2)

while len(lista1) != 0:
  nowy = lista1.pop(0)
  lista2.append(nowy)
```

```python
#zad3 

lista1 = ["Pomidor", "Czosnek", "Oliwa", "", "Makaron"]
lista2 = []
index = 0

while len(lista1) != 0:
  if lista1[index] != "":
    nowy = lista1.pop(0)
    lista2.append(nowy)
  else:
    break
index += 1


print(lista1)
print(lista2)
```

```python
#zad3 Napisz program, który przyjmie listę [1, 2, 3, 4, 5].
# W wyniku zwróci wszystkie wartości w liście pomnożone o liczbe podaną przez usera.

x1 = [1, 2, 3, 4, 5]
liczba = input("Podaj liczbę: ")

for element in x1:
  print(f"{element} * {liczba} równa się {element * int(liczba)}")
```

```python
#4 Napisz program, który przyjmie od użytkownika dwie liczby.
# Nastepnie wyświetli użytkownikowi wszystkie liczby znajdujące się
#pomiędzy tymi liczbami, np dla n1 = 0 n2 = 5 wyswietli 0, 1, 2, 3, 4, 5

n1 = int(input("Podaj liczbe 1: "))
n2 = int(input("Podaj liczbe 2: "))+1
liczba = int(n1)

while liczba != n2:
  print(liczba)
  liczba += 1
```

```python
n1 = int(input("Podaj liczbe 1: "))
n2 = int(input("Podaj liczbe 2: ")) + 1

for i in range(n1, n2):
  print(i)
```

```python
#6 Napisz program, który poda sumę wszystkich n liczb, gdzie n
# to liczba podana przez usera. np dla n = 3 wykona działanie 0 + 1 + 2 + 3

n = int(input("Podaj liczbe: ")) + 1
suma = 0

for i in range(n):
  suma += i
print(suma)

```
